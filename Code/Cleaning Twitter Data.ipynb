# Cleaning

```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.feature_extraction import stop_words
from nltk.stem.porter import PorterStemmer

from bs4 import BeautifulSoup 

import datetime as dt
```

```python
df_WI = pd.read_csv('../Data/df_WI.csv')
```

```python
df_WI
```

```python
df_WI.info()
```

```python
df_TX = pd.read_csv('../Data/df_TX.csv')
df_TX.head()
```

```python
df_concat = pd.concat([df_WI, df_TX], ignore_index = True)
```

```python
df_concat.info()
```

```python
# drop Unanamed: 0, geolocation, User, user_split, Text columns
df_concat.drop(columns = ['Unnamed: 0', 'Geolocation', 'User'], inplace = True)
```

```python
df_concat
```

```python
# Change Date column to datetime
df_concat['Date'] = pd.to_datetime(df_concat['Date'])
```

```python
df_concat['Date'].dtypes
```

```python
# Analyze Nulls
df_concat.isnull().sum().sort_values(ascending=False)
```

```python
df_concat['Mentions'].fillna('None', inplace = True)
df_concat['HashTags'].fillna('None', inplace = True)
```

```python
df_concat.shape
```

```python
df_concat['Text'].isnull().sum()
```

```python
df_concat.drop(index = df_concat[df_concat['Text'].isnull()].index, inplace = True)
```

```python
df_concat.info()
```

```python
df_concat.to_csv('../Data/df_TX_WI_clean.csv')
```
