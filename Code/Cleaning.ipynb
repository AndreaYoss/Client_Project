## Manual Classification

### Iowa Data

Splitting the Iowa data into seperate y=1 and y=0 dfs.

```python
df_IA_1 = df_IA[df_IA['target']==1]
df_IA_0 = df_IA[df_IA['target']==0]
```

Starting with y=1 target in the Iowa df:

```python
df_IA_1.head()
```

Creating a y=0 column as a 'check' in the y=1 df.

```python
df_IA_1['check'] = 0
```

Inspecting individual user tweeting frequencies as starting point for analysis.

```python
df_IA_1['user_split'].value_counts().head()
```

Through inspection and our domain knowlegde, we see that users with 'NWS', 'iembot', 'NWS', or 'CP' in their username appear to be associated with y=1, so we can change the value of the check for these from 0 to 1.

```python
df_IA_1['check'][df_IA_1[df_IA_1['user_split'].str.contains('NWS')].index] = 1
df_IA_1['check'][df_IA_1[df_IA_1['user_split'].str.contains('iembot')].index] = 1
df_IA_1['check'][df_IA_1[df_IA_1['user_split'].str.contains('USGS')].index] = 1
df_IA_1['check'][df_IA_1[df_IA_1['user_split'].str.contains('CP')].index] = 1
```

```python
df_IA_1['check'].value_counts(normalize=True)
```

*Note that approximately 75% of our y=1 data is from users with 'NWS', 'iembot', 'NWS', or 'CP' in their names.*

Based on domain knowlege, we compiled a list of specific strings associated with true y=1 targets. Note that our list is case sensitive to ensure. 

```python
flood_1 = ['flood warning', 'intersection', 'flood waters', 'flood (warning)', 'traffic',
          'edu', 'flash']
```

```python
for word in flood_1:
    for i in df_IA_1.index:
        col = df_IA_1['Text'][i].lower()
        if col.find(word)>-1:
            df_IA_1['check'][i]=1
```

```python
df_IA_1['check'].value_counts()
```

Manual Inspection:  
We will identify any true y=1.

```python
df_IA_1[df_IA_1['check']==0].index
```

```python
df_IA_1[df_IA_1['check']==0]['Text'][55]
```

```python
ind_1 = [41,42,55, 84,86, 87, 121, 122, 144, 150, 151, 216, 217, 220, 221, 
         224, 225, 286, 303, 467, 485, 543, 589, 606, 618, 620,629]
df_IA_1['check'][ind_1]=1
```

```python
df_IA_1['check'].value_counts(normalize=True)
```

Initially 18.9% of our y=1 targets were false positives. We will fix this by reassigning these targets to y=0, and dropping the 'check' column.

```python
df_IA_1['target'] = df_IA_1['check']
df_IA_1.drop(columns = ['check'], inplace = True)
```

Inspecting initial y=0 case, IA df:

Note: Since GetOldTweets scraper is not very precise with respect to state boundaries and we know from the FEMA dataset that there wasn't a severe flood in Iowa in the scraped time period, I will remove any tweets referring to an actual flood outside of Iowa.

```python
df_IA_0.head()
```

Our top users have the same attributes that we previously determined were indicative of a flood: 'iembot,' 'NWS,' and 'USGS.' Through manual inspection, we see that these do warn of actual floods; however, they warn of floods outside of Iowa, and therfore, we will remove these rows from the dataframe, as we are assuming that in a production we will have specific geolocation data.

```python
df_IA_0.drop(index = df_IA_0[df_IA_0['user_split'].str.contains('iembot')].index, inplace = True)
df_IA_0.drop(index = df_IA_0[df_IA_0['user_split'].str.contains('USGS')].index, inplace = True)
df_IA_0.drop(index = df_IA_0[df_IA_0['user_split'].str.contains('NSW')].index, inplace = True)
df_IA_0.drop(index = df_IA_0[df_IA_0['user_split'].str.contains('CP')].index, inplace = True)
```

Creating and exporting an 'adjusted' Iowa dataframe.

```python
df_IA_adjusted = pd.concat([df_IA_1, df_IA_0], ignore_index = True)
df_IA_adjusted.head()
```
